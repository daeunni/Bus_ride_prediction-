```python
import pandas as pd
import numpy as np
import os

folder = 'C:/Users/이다은/Desktop/BOAZ 14기/데이콘 프로젝트^ㅁ^/버스 승차량 예측/ensemble'

nf = 0
for f in os.listdir(folder):
    ext = os.path.splitext(f)[-1]
    if ext == '.csv': 
        s = pd.read_csv(folder+"/"+f)
    else: 
        continue
    if len(s.columns) !=2:
        continue
    if nf == 0: 
        slist = s
    else: 
        slist = pd.merge(slist, s, on="id")
    nf += 1
```

```python
# 이 파라미터는 멱평균 앙상블에 있어 중요한 수치임. 최적의 수치를 찾기 바랍니다.    
p = 0.5

if nf >= 2:
    pred = 0
    for j in range(nf): pred = pred + slist.iloc[:,j+1]**p 
    pred = pred / nf    
    pred = pred**(1/p)

    submit = pd.DataFrame({'id': slist.id, '18~20_ride': pred})
    t = pd.Timestamp.now()
    fname = "2번째0.5.csv"
    submit.to_csv(fname, index=False)
    
    print(fname)
```
